```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df[['Survived','Pclass','Age','Fare']], hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[['Survived','Pclass','Age','Fare']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(), bins=30)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Fare'], bins=30)
plt.title("Histogram of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Boxplot of Age vs Survival")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Boxplot of Fare vs Survival")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatterplot of Age vs Fare")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Fare', y='Age', hue='Survived', data=df,color='red')
plt.title("Scatterplot of Age vs Fare by Survival")
plt.show()
```
